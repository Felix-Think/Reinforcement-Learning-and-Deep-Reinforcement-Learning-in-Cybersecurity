```python
import matplotlib.pyplot as plt
import numpy as np
def smooth(scalars, weight=0.85):
    """Làm mượt biểu đồ (Exponential Moving Average)"""
    if not scalars: return []
    last = scalars[0]
    smoothed = []
    for point in scalars:
        smoothed_val = last * weight + (1 - weight) * point
        smoothed.append(smoothed_val)
        last = smoothed_val
    return smoothed

results = {
    "CartPole-v1": {
        "AWR": (CP_AWR_train_rewards, CP_AWR_test_rewards),
        "PPO": (CP_PPO_train_rewards, CP_PPO_test_rewards)
    },
    "Acrobot": {
        "AWR": (AC_AWR_train_rewards, AC_AWR_test_rewards), 
        "PPO": (AC_PPO_train_rewards, AC_PPO_test_rewards)
    },
    "LunarLander": {
        "AWR": (LL_AWR_train_rewards, LL_AWR_test_rewards), 
        "PPO": (LL_PPO_train_rewards, LL_PPO_test_rewards)
    }

}
def plot_benchmarks(results_dict):
    envs = list(results_dict.keys())
    fig, axes = plt.subplots(len(envs), 2, figsize=(16, 5 * len(envs)))
    fig.suptitle('Benchmark: PPO vs AWR', fontsize=20, fontweight='bold', y=0.95 + (3 - len(envs))*0.02)
    
    # Nếu chỉ có 1 môi trường, axes sẽ là mảng 1 chiều, ta chuyển về 2 chiều để dễ xử lý
    if len(envs) == 1: axes = np.array([axes])
    
    colors = {'PPO': '#1f77b4', 'AWR': '#ff7f0e'}
    
    for i, env_name in enumerate(envs):
        for algo in ['PPO', 'AWR']:
            if algo in results_dict[env_name]:
                train_rew, test_rew = results_dict[env_name][algo]
                
                # Cột 1: Train Rewards
                axes[i, 0].plot(train_rew, alpha=0.2, color=colors[algo])
                axes[i, 0].plot(smooth(train_rew), label=f'{algo}', color=colors[algo], linewidth=2)
                axes[i, 0].set_title(f'{env_name} - Train Rewards', fontsize=14)
                axes[i, 0].set_ylabel('Reward')
                axes[i, 0].grid(True, alpha=0.3)
                
                # Cột 2: Test Rewards
                axes[i, 1].plot(test_rew, alpha=0.2, color=colors[algo])
                axes[i, 1].plot(smooth(test_rew), label=f'{algo}', color=colors[algo], linewidth=2)
                axes[i, 1].set_title(f'{env_name} - Test Rewards', fontsize=14)
                axes[i, 1].grid(True, alpha=0.3)
                
        axes[i, 0].legend()
        axes[i, 1].legend()
        
        if i == len(envs) - 1:
            axes[i, 0].set_xlabel('Episodes')
            axes[i, 1].set_xlabel('Episodes')
    plt.tight_layout(rect=[0, 0.03, 1, 0.93])
    plt.show()
# Gọi hàm để vẽ
plot_benchmarks(results)
```
